```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format= '%d/%m/%Y')
```

```python
df.info()
```

```python
orders_placed = df['Order ID'].nunique()
orders_placed
```

```python
orders_shipped = df[df['Ship Date'].notna()]['Order ID'].nunique()
orders_shipped
```

```python
orders_by_segment = df.groupby('Segment')['Order ID'].count().reset_index()
orders_by_segment
```

```python
orders_by_segment.columns = ['Segment', 'Orders']
orders_by_segment
```

```python
sales_by_segment = df.groupby('Segment')['Sales'].sum().reset_index()
sales_by_segment
```

```python
sales_by_categories = df.groupby(['Category', 'Sub-Category'])['Sales'].sum().reset_index()
sales_by_categories
```

```python
sales_by_categories.to_csv('sales_by_categories.csv', index=False)
```

```python
sales_by_segment.to_csv('sales_by_segemnt.csv', index=False)
```
